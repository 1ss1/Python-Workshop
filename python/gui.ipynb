```python
from tkinter import *
from tkcalendar import DateEntry

#PART- I

master = Tk() # arayüz ismi (parent)

canvas = Canvas(master, height=450, width=750) # parrent object

canvas.pack()   #  arayüzün üzerine birşey yerleştirmeye yarayan komutlar:
                # pack
                # place
                # grid
            
frame_ust = Frame(master, bg='#add8e6')
frame_ust.place(relx=0.1, rely=0.1, relwidth=0.8, relheight=0.1) # relx ve rely başlangıç noktasına uzaktık 
                                                                  # % 10 luk mesafeden başla dedik.
                                                                  # %10 dan başladıktan sonra %80 lik uzunluğa ulaşacak.

frame_alt_sol = Frame(master, bg='#add8e6')
frame_alt_sol.place(relx=0.1, rely=0.21, relwidth=0.23, relheight=0.5) # aradaki minik %1 lik boşluk için %21 dedik.
        
frame_alt_sağ = Frame(master, bg='#add8e6')
frame_alt_sağ.place(relx=0.34, rely=0.21, relwidth=0.56, relheight=0.5)   

hatırlatma_tipi_etiket = Label(frame_ust, bg='#add8e6', text= "Hatırlatma Tipi:", font="Verdana 12 bold")  
# stile dair işlemler label içinde yapılır.

hatırlatma_tipi_etiket.pack(padx=10, pady=10, side=LEFT) # padx,y = padding   side=left, right, bottom, top yasslamaa için kullanılır
# konuma dair değişiklikler pack içinde yapılır 

hatırlatma_tipi_opsiyon = StringVar(frame_ust)
hatırlatma_tipi_opsiyon.set("\t") # default değer

hatırlatma_tipi_acilir_menu = OptionMenu(frame_ust,
                                         hatırlatma_tipi_opsiyon,
                                         "Doğum Günü",
                                         "Alışveriş",
                                         "Ödeme")
hatırlatma_tipi_acilir_menu.pack(padx=10, pady=10, side=LEFT)


# dateEntry tkinter in altında olmadığı için en üstteki yüklemeyi yaptık
hatırlatma_tarihi_secici = DateEntry(frame_ust, width=12, background='orange', foreground= 'black', borderwidth=1, locale='de_DE') 
hatırlatma_tarihi_secici._top_cal.overrideredirect(False)
hatırlatma_tarihi_secici.pack(padx=10, pady=10, side=RIGHT)
# https://tkcalendar.readthedocs.io/en/stable/DateEntry.html background gibi ayrıntıları kopyaladığımız site

hatırlatma_tarihi_etiket = Label(frame_ust, bg='#add8e6', text= "Hatırlatma Tarihi:", font="Verdana 12 bold")  
hatırlatma_tarihi_etiket.pack(padx=10, pady=10, side=RIGHT) 
# side ların doğru sıralanması için etiket, acılır menü ve seçici nin sıralarına  dikkat

# PART -II

Label(frame_alt_sol, text="Hatırlatma Yöntemi", bg='#add8e6', font="Verdana 10 bold").pack(padx=10, pady=10, anchor=NW) # NW= North-West
# değişken ismi vermediğimiz için Label e pack ekleyebildik.

var = IntVar()

R1 = Radiobutton(frame_alt_sol, text="Sisteme Kaydet", variable=var, value=1, bg='#add8e6', font="Verdana 10")
R1.pack(anchor=NW, pady=5, padx=15)

R2 = Radiobutton(frame_alt_sol, text="E-posta Gönder", variable=var, value=2, bg='#add8e6', font="Verdana 10")
R2.pack(anchor=NW, pady=5, padx=15)

var1 = IntVar()
C1 = Checkbutton(frame_alt_sol, text="Bir Hafta Önce", variable=var1, onvalue= 1, offvalue= 0, bg='#add8e6', font="Verdana 10") # onvalue= seçili olduğu zaman alacağı değer offvalue= seçili olmalığı zaman alacağı değer 
C1.pack(anchor=NW, pady=2, padx=25)

var2 = IntVar()
C2 = Checkbutton(frame_alt_sol, text="Bir Gün Önce", variable=var2, onvalue= 1, offvalue= 0, bg='#add8e6', font="Verdana 10")
C2.pack(anchor=NW, pady=2, padx=25)

var3 = IntVar()
C3 = Checkbutton(frame_alt_sol, text="Aynı Gün", variable=var3, onvalue= 1, offvalue= 0, bg='#add8e6', font="Verdana 10")
C3.pack(anchor=NW, pady=2, padx=25)
# Checkbutton lar eş zamanlı seçilebildiği için 

# PART-III

from tkinter import messagebox

def gonder():
    son_mesaj = ""
    
    try:
        if var.get():  # sisteme kaydet butonuna basıldığından emin olmak için yazdık
            if var.get() == 1:
                son_mesaj += "Veriniz başarıyla sisteme kaydedilmiştir."

                tip = hatırlatma_tipi_opsiyon.get() if  hatırlatma_tipi_opsiyon.get() == '' else "Genel"
                tarih = hatırlatma_tarihi_secici.get() 
                mesaj = metin_alani.get("1.0", "end") # get() tek başına text te çalışmadığı için başlangıç-bitiş noktası verdik

                with open("hatırlatmalar.txt", "w") as dosya:
                    dosya.write('{} katagorisinde, {} tarihinde ve "{}" notuyla hatırlatma'.format(tip, tarih, mesaj))
                    dosya.close()

            elif var.get() == 2:
                son_mesaj += "E-posta yoluyla hatırlatma size ulaşacaktır."

            messagebox.showinfo("Başarılı İşlem", son_mesaj)

        else:
            son_mesaj += "Gerekli alanların doldurulduğundan emin olun!"
            messagebox.showwarning("Yetersiz Bilgi", son_mesaj)
            
    except:
        son_mesaj += "İşlem başarısız oldu!"
        messagebox.showerror("Yetersiz Bilgi", son_mesaj)
            
    finally:
        master.destroy()
    
    return

Label(frame_alt_sağ, text="Hatırlatma Mesajı", bg='#add8e6', font="Verdana 10 bold").pack(padx=10, pady=10, anchor=NW) # NW= North-West

metin_alani = Text(frame_alt_sağ, height= 9, width= 50)
metin_alani.tag_configure('style', foreground = '#bfbfbf', font=('Verdana', 10, 'bold'))
metin_alani.pack()
karsilama_metni = 'Mesajınzı buraya giriniz...'
metin_alani.insert(END, karsilama_metni, 'style')

gonder_butonu = Button(frame_alt_sağ, text="Gönder", command=gonder) # command bir fonksiyon. tetiklenmesi beklenen aksiyon
                                                                     # command=master.destroy da olabilir.

gonder_butonu.pack(anchor=S) # S = south

master.mainloop() # sürekli bir döngü içinde çaalışır. Biz arayüz ekranını kapatana kadar hücre çalışmaya devam eder.
```
